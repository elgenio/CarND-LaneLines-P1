```python
batch = [[1/16,2/16,1/16],
         [2/16,4/16,2/16],
         [1/16,2/16,1/16]]
```

```python
def applyBatch(batch,image,i,j,f):
    x = 0
    for k in range(0,2):
        for h in range(0,2):
            x = x + batch[k][h]*image[i+k-1][j+h-1][f]
    return x
```

```python
for f in range(0,2):
    for i in range(1,image.shape[0]):
        for j in range(1,image.shape[1]):
            image[i][j][f]=applyBatch(batch,image,i,j,f)
```
